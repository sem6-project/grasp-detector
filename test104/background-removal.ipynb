```python
from matplotlib import pyplot as plt
import utils
```

```python
dps = utils.prepare_datapoints()
bgs = utils.read_backgrounds()
bg_mapping = utils.get_background_mappings()
```

```python
d = dps[100]
bg = bgs[bg_mapping[d.image_name]]
img = d.get_image(gray=True, background=None)
bg = utils.cv2.cvtColor(bg, utils.cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img)
```

```python
d.background_image = bg
x = d.get_image(background=bg)
plt.imshow(x)
```

```python
plt.imshow(bg)
```

```python
diff = (img.astype('float') - bg.astype('float')).astype('uint8')
plt.imshow(diff)
```

```python
absdiff = utils.cv2.absdiff(img, bg)
plt.imshow(absdiff)
```

```python
median = utils.cv2.medianBlur(diff, 7)
plt.imshow(median)
```

```python
second_median = utils.cv2.medianBlur(median, 5)
plt.imshow(second_median)
```
